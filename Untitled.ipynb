```python
import requests
from lxml import etree
from urllib.parse import urljoin

def load_page(url):
    response=requests.get(url)
    response.encoding=response.apparent_encoding
    html=response.text
    return html


def parse_new(html):
    new={}
    try:
        s=etree.HTML(html)
        title=s.xpath("//h2/text()")[0]
        author=s.xpath("//div[@class='zz'][1]/text()")[0].split()[0].lstrip("作者：")
        source=s.xpath("//div[@class='zz'][1]/text()")[0].split()[1].lstrip("来源：")
        time_fb=s.xpath("//div[@class='zz']/text()")[1].strip().lstrip("发布时间：").rstrip("  浏览：").strip()
        content=s.xpath("//div[@id='vsb_content']//p/text()")
        acticle="".join(x.strip() for x in content)
        # print(title)
        # print(author)
        # print(source)
        # print(time_fb)
        # # print(content)
        # print(acticle)
        new['title']=title
        new['author']=author
        new['source']=source
        new['time_fb']=time_fb
        new['content']=acticle
    except Exception as e:
            print(e)
    return new

def urls_parse(url,num):
    urls=[]
    for i in range(0,num):
        s=etree.HTML(load_page(url))
        result=s.xpath("//div[@class='right-1']//li/a/@href")
#         print(result)
        for u in result:
            urls.append(urljoin("http://www.zjitc.net",u))
            
        page_next=s.xpath("//a[@class='Next'][1]/@href")[0]
        
        if page_next.startswith("xyxw"):
            url=urljoin("http://www.zjitc.net/xwzx/",page_next)
        else:
            url=urljoin("http://www.zjitc.net/xwzx/xyxw/",page_next)
#         print(url)
    return urls
            
if __name__=="__main__":
    url="http://www.zjitc.net/xwzx/xyxw.htm"
    num=int(input("请输入爬取的页数"))
    news=[]
    urls=urls_parse(url,num)
    print(len(urls))
    print(len(list(set(urls))))
    for u in list(set(urls)):
        news.append(parse_new(load_page(u)))
    with open("a.txt","w",encoding="utf-8") as file:
          for i in news:
              file.write(str(i)+"\n")
#     with open("b.txt","w",encoding="utf-8") as file:
#         file.write(str(news))
```
